### 2. Examine the State and Action Spaces

In this environment, two agents control rackets to bounce a ball over a net. If an agent hits the ball over the net, it receives a reward of +0.1.  If an agent lets a ball hit the ground or hits the ball out of bounds, it receives a reward of -0.01.  Thus, the goal of each agent is to keep the ball in play.

The observation space consists of 8 variables corresponding to the position and velocity of the ball and racket. Two continuous actions are available, corresponding to movement toward (or away from) the net, and jumping. 

Run the code cell below to print some information about the environment.

````python
# reset the environment
env_info = env.reset(train_mode=True)[brain_name]

# number of agents 
num_agents = len(env_info.agents)
print('Number of agents:', num_agents)

# size of each action
action_size = brain.vector_action_space_size
print('Size of each action:', action_size)

# examine the state space 
state = env_info.vector_observations
state_size = state.shape[1]

print('There are {} agents. Each observes a state with length: {}'
      .format(state.shape[0], state_size))
print('The observation for the first agent looks like:\n', state[0])
````

````python
## Initialize values in Config
config = Config()
config.seed = 997
config.device = 'cpu'
config.target_score = 0.5
config.target_episodes = 100

config.num_agents = num_agents
config.action_size = action_size
config.state_size = state_size

config.actor_layers = [64, 64]
config.critic_layers = [64, 64]
config.actor_lr = 0.003
config.critic_lr = 0.0004
config.lr_sched_step = 1
config.lr_sched_gamma = 0.1
config.batch_normalization = True

config.buffer_size = int(1e6)
config.batch_size = 64
config.gamma = 0.99
config.tau = 0.01
config.noise = True
config.noise_theta = 0.9
config.noise_sigma = 0.01

config.max_episodes = 10000

# Instantiate a Multi Agent
maddpg = MultiAgentDDPG()
````

````python
# The training function
def train(env, maddpg, max_episodes=5000):
    """Train a Multi Agent Deep Deterministic Policy Gradients (MADDPG).
    Params
    ======
        env (UnityEnvironment): environment for the agents
        maddpg (MADDPG): the Multi Agent DDPG
        n_episodes (int): maximum number of training episodes
    """
    scores = []
    agents_scores = []
    moving_avg = []
    
    # Checkpoints of target score by which to reduce the learning rate
    avg_checkpoints = iter([0.4, 0.6])
    next_avg_checkpoint = next(avg_checkpoints)
    # Perform training for 'n' number of episodes
    for i_episode in range(1, max_episodes+1):
        # reset noise to initial value
        maddpg.reset_noise()
        # set train_mode to True
        env_info = env.reset(train_mode=True)[brain_name]
        # obtain state from Unity environment
        state = env_info.vector_observations
        # create an array of zeros to store scores for each agent
        scores_episode = np.zeros(config.num_agents)
        while True:
            # Perform a step: S;A;R;S'
            # obtain actions from maddpg class with state values
            actions = maddpg.act(state)
            # send all actions to the environment
            env_info = env.step(actions)[brain_name]
            # obtain rewards as a result of the actions
            rewards = env_info.rewards
            # obtain the next_state based on actions executed in the environment
            next_state = env_info.vector_observations

            # store the experience tuple in the replay buffer and update the actor & critic weights
            maddpg.store(state, actions, rewards, next_state)

            # update state and scores_episode for the next iteration
            state = next_state
            scores_episode += rewards

            if any(env_info.local_done):
                break
        
        # append data to list
        agents_scores.append(scores_episode)
        scores.append(scores_episode.max())
        moving_avg.append(np.mean(scores[-config.target_episodes:]))

        # Decrease learning rate if last value of moving average is more ot equivalent to next_avg_checkpoint value
        if moving_avg[-1] >= next_avg_checkpoint:
            print('\nDecreasing learning rate ...')
            maddpg.lr_step()
            next_avg_checkpoint = next(avg_checkpoints)

        print('\rEpisode {:4d}\t' \
              'Last score: {:5.2f} ({:5.2f} / {:5.2f})\t' \
              'Moving average: {:5.3f}'
              .format(i_episode,
                      scores[-1], scores_episode[0], scores_episode[1],
                      moving_avg[-1]),
              end='')
        if i_episode % 100 == 0:
            print()

        # Check if the environment has been solved, if yes terminate training and print score.
        if moving_avg[-1].mean() >= config.target_score \
                and i_episode >= config.target_episodes:
            print('\n\nEnvironment solved in {:d} episodes!\t' \
                  'Moving Average Score: {:.3f}'
                  .format(i_episode-config.target_episodes, moving_avg[-1]))
            break

    print('\n')
    return scores, agents_scores, moving_avg
````

````python
## Train the agent
print('Starting training phase using these settings:')
print("\n '".join(str(config.__dict__).split(", '")), '\n')

%time scores, agents_scores, moving_avg = train(env, maddpg, config.max_episodes)
````

````python
# Plot graph of rewards

# Trace a line to indicate the target value
target = [config.target_score] * len(scores)

# Graph of the total rewards
fig = plt.figure(figsize=(18,8))
ax = fig.add_subplot(111)
ax.set_title('Plot of rewards', fontsize='xx-large')
ax.plot(scores, label='Score', color='Blue')
ax.plot(moving_avg, label='Moving Average', color='Red', linewidth=3)
ax.plot(target, linestyle='--', color='Green', linewidth=1 )
ax.text(0, config.target_score, 'Target', color='Black', fontsize='large')
ax.set_ylabel('Score')
ax.set_xlabel('Episode No.')
ax.legend(fontsize='xx-large')
plt.show()

# Graph for each one of the agents
fig, axs = plt.subplots(1, 2, figsize=(18, 8), sharex=True, sharey=True)
fig.suptitle('Rewards for each agents', fontsize='xx-large')
axs = axs.flatten()
for idx, (ax, s) in enumerate(zip(axs, np.transpose(agents_scores))):
    ax.plot(s, label='Agent Score', color='Turquoise', zorder=1)
    ax.plot(moving_avg, label='Moving Average (Total)', color='Orange', linewidth=3, alpha=0.655, zorder=2)
    ax.plot(target, linestyle='--', color='Silver', linewidth=1, zorder=0)
    ax.text(0, config.target_score, 'Target', color='Black', fontsize='large')
    ax.set_title('Agent #%d' % (idx+1))
    ax.set_ylabel('Score')
    ax.set_xlabel('Episode No.')
    ax.label_outer()
    ax.legend(fontsize='medium')
plt.show()
````

````python
checkpoint = {
    'maddpg_state_dict': maddpg.state_dict(),
    'scores': scores,
    'agents_scores': agents_scores,
    'moving_avg': moving_avg}
torch.save(checkpoint, 'checkpoint.pt')
````

````python
# Test the trained model
def test(env, maddpg, max_episodes=3):
    """Test a Multi Agent Deep Deterministic Policy Gradients (MADDPG).
    Params
    ======
        env (UnityEnvironment): environment for the agents
        maddpg (MADDPG): the Multi Agent DDPG
        n_episodes (int): maximum number of training episodes
    """
    # Perform training for 'n' number of episodes
    for i_episode in range(1, max_episodes+1):
        # Reset the environment
        env_info = env.reset(train_mode=False)[brain_name]
        # Initialize numpy zero array to save agent scores
        scores = np.zeros(config.num_agents)
        while True:
            # Obtain agent states from environment
            states = env_info.vector_observations
            # Obtain actions for each agent from MADDPG agent class based on current states
            actions = maddpg.act(states)
            # Send all actions to the environment
            env_info = env.step(actions)[brain_name]
            # Update the score (for each agent)
            scores += env_info.rewards
            # Obtain next state (for each agent)
            states = env_info.vector_observations
            # Exit loop if episode finished
            if any(env_info.local_done):
                break
        print('\rEpisode {:4d}\tScore: {:5.2f} ({:5.2f} / {:5.2f})\t'
              .format(i_episode, scores.max(), scores[0], scores[1]))

# Load checkpoint
checkpoint = torch.load('checkpoint.pt')
# Set noise to False
config.noise = False
# Instantiate an object of MADDPG class
maddpg = MultiAgentDDPG()
# Pass the checkpoint maddpg_state_dict to maddpg's load_state_dict 
maddpg.load_state_dict(checkpoint['maddpg_state_dict'])
# Run test on the model based on saved training
test(env, maddpg)
````
